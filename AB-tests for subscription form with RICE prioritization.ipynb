# Определение наиболее перспективной (оценка RICE) гипотезы для увеличения выручки интернет-магазина, A/B-тестирование и оценка бизнес-метрик

Совместно с отделом маркетинга интернет-магазина был подготовлен список гипотез для увеличения выручки.

Необходимо:
- приоритизировать гипотезы;
- определить, есть ли значимые различия в показателях, влияющих на выручку, между группами A и B.

**Описание данных**

- гипотезы по увеличению выручки интернет-магазина:
    - `Hypothesis` — краткое описание гипотезы;
    - `Reach` — охват пользователей по 10-балльной шкале;
    - `Impact` — влияние на пользователей по 10-балльной шкале;
    - `Confidence` — уверенность в гипотезе по 10-балльной шкале;
    - `Efforts` — затраты ресурсов на проверку гипотезы по 10-балльной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.
    

- данные заказов:
    - `transactionId` — идентификатор заказа;
    - `visitorId` — идентификатор пользователя, совершившего заказ;
    - `date` — дата, когда был совершён заказ;
    - `revenue` — выручка от заказа;
    - `group` — группа A/B-теста, в которую попал заказ.
    

- таблица с количеством посетителей в указанные даты:
    - `date` — дата;
    - `group` — группа A/B-теста;
    - `visitors` — количество пользователей в указанную дату в указанной группе A/B-теста.

```python
# импортируем нужные библиотеки
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
from IPython.display import display
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
```

## Приоритизация гипотез

В файле '/datasets/hypothesis.csv' гипотезы по увеличению выручки интернет-магазина с указанными параметрами Reach, Impact, Confidence, Effort.

```python
data = pd.read_csv('/datasets/hypothesis.csv') # запишем файл в переменную
```

```python
data # визуально изучим файл
```

### Фреймворк ICE

```python
data['ICE'] = data['Impact'] * data['Confidence'] / data['Efforts']
```

```python
data[['Hypothesis', 'ICE']].sort_values('ICE', ascending=False)
```

### Фреймворк RICE

```python
data['RICE'] = (data['Reach'] * data['Impact'] * data['Confidence']) / data['Efforts']
```

```python
data[['Hypothesis', 'RICE']].sort_values('RICE', ascending=False)
```

### Выводы

Чтобы максимизировать ожидаемое улучшение ключевых метрик бизнеса, необходимо выбрать наиболее перспективные гипотезы. Согласно RICE, наибольший приоритет с отрывом по баллам от второго места в 2 раза имеет гипотеза №7:

```python
data['Hypothesis'][7]
```

В обеих модификациях способов приоритизации среди пятерки гипотез-лидеров присутствуют 0, 7, 6, 2, 8, но гипотезы на разных местах, так как модификация RICE учитывает охват пользователей, а ICE - нет.

Предпочтительнее ориентироваться на результаты RICE для выбора порядка проверки гипотез, так как важно учитывать, сколько пользователей затронет гипотеза.

### Общая информация о группах теста

```python
print('Начало теста:', visitors['date'].min())
print('Окончание теста:', visitors['date'].max())
```

```python
print('Количество пользователей в группе A по таблице с заказами:', orders[orders['group'] == 'A']['visitorId'].nunique())
print('Количество пользователей в группе B по таблице с заказами:', orders[orders['group'] == 'B']['visitorId'].nunique())
```

не попадают ли какие-то пользователи в обе группы - по таблице с заказами

```python
buyers_A_id = orders[orders['group'] == 'A']['visitorId']
buyers_B_id = orders[orders['group'] == 'B']['visitorId']
```

```python
intersection_users = pd.Series(list(set(buyers_A_id).intersection(set(buyers_B_id))))
```

```python
print('В обе группы попадают', len(intersection_users), 'пользователей.')
```

Удалим данные пользователей, попавших в обе группы:

```python
orders = orders.query('visitorId not in @intersection_users')
```

### График кумулятивной выручки по группам

```python
# создадим массив уникальных пар значений дат и групп
dates_groups = orders[['date','group']].drop_duplicates() 
```

```python
orders_agg = dates_groups.apply(
    lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])].agg(
        {
            'date' : 'max', 
            'group' : 'max', 
            'transactionId' : 'nunique', 
            'visitorId' : 'nunique', 
            'revenue' : 'sum'
        }
    ), axis=1
).sort_values(by=['date','group']) # кумулятивные данные по заказам

visitors_agg = dates_groups.apply(
    lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])].agg(
        {
            'date' : 'max', 
            'group' : 'max', 
            'visitors' : 'sum'
        }
    ), axis=1
).sort_values(by=['date','group']) # кумулятивные данные по пользователям

# объединяем кумулятивные данные в одной таблице
cumulative_data = orders_agg.merge(visitors_agg, on=['date', 'group'], how='outer')
cumulative_data.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
```

```python
# датафреймы с кумулятивными данными по дням для групп A и B
cumulative_revenue_A = cumulative_data[cumulative_data['group']=='A'][['date','revenue', 'orders']]
cumulative_revenue_B = cumulative_data[cumulative_data['group']=='B'][['date','revenue', 'orders']]
```

```python
# графики кумулятивной выручки
plt.figure(figsize=(10, 5))
plt.title(label='График кумулятивной выручки по группам')
plt.plot(cumulative_revenue_A['date'], cumulative_revenue_A['revenue'], label='A')
plt.plot(cumulative_revenue_B['date'], cumulative_revenue_B['revenue'], label='B')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Кумулятивная выручка', fontsize=12)
plt.show()
```

#### Выводы и предположения

Выручка увеличивается в течение всего теста, что является позитивным признаком - пользователи совершают новые и новые заказы.

Однако графики выручки обеих групп в нескольких точках резко растут. Это может сигнализировать о всплесках числа заказов, либо о появлении очень дорогих заказов в выборке.

### График кумулятивного среднего чека по группам

```python
plt.figure(figsize=(10, 5))
plt.title(label='График кумулятивного среднего чека по группам')
plt.plot(cumulative_revenue_A['date'], cumulative_revenue_A['revenue'] / cumulative_revenue_A['orders'], label='A')
plt.plot(cumulative_revenue_B['date'], cumulative_revenue_B['revenue'] / cumulative_revenue_B['orders'], label='B')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Кумулятивный средний чек', fontsize=12)
plt.show()
```

#### Выводы и предположения

Средний чек также подвержен скачкам и становится более равномерным ближе к концу теста: относительно стабилизируется для группы A и показывает тенденцию к снижению для группы B. 

Возможно, скачок графика для группы B в середине теста обусловлен крупным заказом, поэтому для группы B требуется больше данных, чтобы прийти к реальному среднему чеку и установиться на его уровне.

### График относительного изменения кумулятивного среднего чека группы B к группе A

```python
cumulative_revenue = cumulative_revenue_A.merge(cumulative_revenue_B, on='date', how='outer', suffixes=['_A', '_B'])
av_check_A = cumulative_revenue['revenue_A'] / cumulative_revenue['orders_A']
av_check_B = cumulative_revenue['revenue_B'] / cumulative_revenue['orders_B']
```

```python
# отношение средних чеков
plt.figure(figsize=(10, 5))
plt.title(label='График относительного изменения кумулятивного среднего чека группы B к группе A')
plt.plot(cumulative_revenue['date'], av_check_B / av_check_A - 1)
plt.axhline(y=0, color='black', linestyle='--') # ось на уровне 0
plt.grid()
plt.xticks(rotation=45)
plt.xlabel('Дата', fontsize=12)
plt.ylabel('Относительное изменение', fontsize=12)
plt.show()
```

#### Выводы и предположения

В нескольких точках график относительного изменения кумулятивного среднего чека между группами резко «прыгает». Это сигнализирует о наличии крупных заказов и выбросов в данных. Отношение среднего чека между группами ещё не установилось.

### График кумулятивной конверсии по группам

## Проверка гипотез

Планируется протестировать 4 гипотезы, в связи с чем необходимо применить поправку на множественную проверку гипотез. Используем поправку Бонферрони. Принять альтернативную гипотезу, если:


<center>$p ≥ \alpha / n, $</center>


где $n$ — количество тестируемых гипотез.

Применяем поправку для уровня значимости 0.05 и четырех гипотез:

<center>0.05 / 4 = 0.0125</center>

### Различия в конверсии между группами по «сырым» данным

***H₀: различий в конверсии между группами A и B по «сырым» данным нет.***

***H₁: различия в конверсии между группами A и B по «сырым» данным есть.***

Уровень статистической значимости α (с учетом поправки Бонферрони): 0.0125
 
Метод проверки гипотез: U-критерий Манна-Уитни (выбросы сдвигают параметрические результаты, поэтому используем параметрический тест).

```python
orders_A = (
    orders[orders['group'] == 'A']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
) # таблица пользователь - количество заказов для группы A
orders_A.columns = ['user_id', 'orders']

orders_B = (
    orders[orders['group'] == 'B']
    .groupby('visitorId', as_index=False)
    .agg({'transactionId': pd.Series.nunique})
) # таблица пользователь - количество заказов для группы B
orders_B.columns = ['user_id', 'orders'] 
```

```python
sample_A = pd.concat(
    [
        orders_A['orders'], pd.Series(0, index=np.arange(
            visitors[visitors['group'] == 'A']['visitors'].sum() - len(orders_A) # нули для не совершавших покупки
        ), name='orders')
    ], axis=0
).sort_values(ascending=False)

sample_B = pd.concat(
    [
        orders_B['orders'], pd.Series(0, index=np.arange(
            visitors[visitors['group'] == 'B']['visitors'].sum() - len(orders_B) # нули для не совершавших покупки
        ), name='orders')
    ], axis=0
).sort_values(ascending=False)
```

```python
# проверим соответствие количества пользователей
print('Сумма длин списков "пользователь-количество заказов" групп A и B:', len(sample_A) + len(sample_B))
print('Сумма количества пользователей по данным датасета visitors:', visitors['visitors'].sum())
```

```python
print('p-value: {0:.3f}'.format(stats.mannwhitneyu(sample_A, sample_B)[1]))
print('Относительное различие между группами: {0:.2%}'.format(sample_B.mean() / sample_A.mean() - 1))
```

#### Выводы и предположения

Согласно «сырым» данным, различие в конверсии между группами A и B есть.

p-value меньше 0.0125. Значит, нулевую гипотезу о том, что статистически значимых различий конверсии между группами нет, отвергаем.

Относительный выигрыш группы B, согласно «сырым» данным, равен 15.98%.

### Различия в среднем чеке заказа между группами по «сырым» данным

***H₀: различий в среднем чеке между группами A и B по «сырым» данным нет.***

***H₁: различия в среднем чеке между группами A и B по «сырым» данным есть.***

Уровень статистической значимости α (с учетом поправки Бонферрони): 0.0125

Метод проверки гипотез: U-критерий Манна-Уитни (выбросы сдвигают параметрические результаты, поэтому используем параметрический тест).

```python
print('p-value: {0:.3f}'.format(
        stats.mannwhitneyu(orders[orders['group'] == 'A']['revenue'], orders[orders['group'] == 'B']['revenue'])[1]
    )
)
print('Относительное различие между группами: {0:.2%}'.format(
    orders[orders['group'] == 'B']['revenue'].mean() / orders[orders['group'] == 'A']['revenue'].mean() - 1)
) 
```

#### Выводы и предположения

Согласно «сырым» данным, различия в среднем чеке между группами A и B нет.

p-value намного больше 0.0125. Значит, нулевую гипотезу о том, что статистически значимых различий среднем чеке между группами нет, мы отвергнуть не можем, относительный выигрыш группы B статистически не значим.

Мы видим, что относительное различие между группами велико, но статистически не значимо - это потому, что тест нечувствителен к выбросам.

### Различия в конверсии между группами по «очищенным» данным

***H₀: различий в конверсии между группами A и B по «очищенным» данным нет.***

***H₁: различия в конверсии между группами A и B по «очищенным» данным есть.***

Уровень статистической значимости α (с учетом поправки Бонферрони): 0.0125

Метод проверки гипотез: U-критерий Манна-Уитни (выбросы сдвигают параметрические результаты, поэтому используем параметрический тест).

```python
many_orders = pd.concat(
    [
        orders_A[orders_A['orders'] > 2]['user_id'],
        orders_B[orders_B['orders'] > 2]['user_id'],
    ],
    axis=0,
) # пользователи с аномально большим количеством заказов

expensive_orders = orders[orders['revenue'] > np.percentile(orders['revenue'], 95)]['visitorId'] # пользователи, совершившие аномально дорогие заказы

abnormal_users = pd.concat([many_orders, expensive_orders], axis=0).drop_duplicates().sort_values() # единый список
```

```python
print('Количество аномальных пользователей:', len(abnormal_users))
```

```python
filtered_sample_A = pd.concat(
    [
        orders_A[np.logical_not(orders_A['user_id'].isin(abnormal_users))]['orders'], pd.Series(0, index=np.arange(
            visitors[visitors['group'] == 'A']['visitors'].sum() - len(orders_A) # нули для не совершавших покупки
        ), name='orders')
    ], axis=0
)

filtered_sample_B = pd.concat(
    [
        orders_B[np.logical_not(orders_B['user_id'].isin(abnormal_users))]['orders'], pd.Series(0, index=np.arange(
            visitors[visitors['group'] == 'B']['visitors'].sum() - len(orders_B) # нули для не совершавших покупки
        ), name='orders')
    ], axis=0
)
```

```python
print('p-value: {0:.3f}'.format(stats.mannwhitneyu(filtered_sample_A, filtered_sample_B)[1]))
print('Относительное различие между группами: {0:.2%}'.format(filtered_sample_B.mean() / filtered_sample_A.mean() - 1))
```

#### Выводы и предположения

Согласно «очищенным» данным, значимое различие в конверсии между группами A и B есть.

Для «очищенных» данных p-value меньше 0.0125 - нулевую гипотезу о том, что статистически значимых различий конверсии между группами нет, отвергаем.

Относительный выигрыш группы B, согласно «очищенным» данным, равен 18.19% - это на 2% больше, чем по «сырым» данным.

### Различия в среднем чеке заказа между группами по «очищенным» данным

***H₀: различий в среднем чеке между группами A и B по «очищенным» данным нет.***

***H₁: различия в среднем чеке между группами A и B по «очищенным» данным есть.***

Уровень статистической значимости α (с учетом поправки Бонферрони): 0.0125

Метод проверки гипотез: U-критерий Манна-Уитни (выбросы сдвигают параметрические результаты, поэтому используем параметрический тест).

```python
print('p-value: {0:.3f}'.format(stats.mannwhitneyu(
    orders[np.logical_and(orders['group'] == 'A', np.logical_not(orders['visitorId'].isin(abnormal_users)))]['revenue'], \
    orders[np.logical_and(orders['group'] == 'B', np.logical_not(orders['visitorId'].isin(abnormal_users)))]['revenue'])[1])
)

print('Относительное различие между группами: {0:.2%}'.format(
    orders[
        np.logical_and(orders['group'] == 'B', np.logical_not(orders['visitorId'].isin(abnormal_users)))
    ]['revenue'].mean() / 
    orders[
        np.logical_and(orders['group'] == 'A', np.logical_not(orders['visitorId'].isin(abnormal_users)))
    ]['revenue'].mean() - 1)
) 
```

#### Выводы и предположения

Согласно «очищенным» от выбросов данным, различия в среднем чеке между группами A и B нет.

p-value намного больше 0.0125, как и для «сырых» данных. Значит, нулевую гипотезу о том, что статистически значимых различий среднем чеке между группами нет, мы отвергнуть не можем.

Относительное различие между группами после удаления выбросов данных упало с почти 28.66% до -4.85%. Разницы в среднем чеке между группами нет, на предыдущие результаты влияли аномальные значения.
